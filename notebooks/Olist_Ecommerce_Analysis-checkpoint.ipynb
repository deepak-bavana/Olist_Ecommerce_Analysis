```python
import pandas as pd
import os

path = r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive"

# Load all 9 files
files = os.listdir(path)
print("Files in folder:")
for f in files:
    print(f)
```

```python
# Load all 9 tables
customers = pd.read_csv(r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive" + r'\olist_customers_dataset.csv')
geolocation = pd.read_csv(r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive" + r'\olist_geolocation_dataset.csv')
orders = pd.read_csv(r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive" + r'\olist_orders_dataset.csv')
order_items = pd.read_csv(r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive" + r'\olist_order_items_dataset.csv')
order_payments = pd.read_csv(r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive" + r'\olist_order_payments_dataset.csv')
order_reviews = pd.read_csv(r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive" + r'\olist_order_reviews_dataset.csv')
products = pd.read_csv(r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive" + r'\olist_products_dataset.csv')
sellers = pd.read_csv(r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive" + r'\olist_sellers_dataset.csv')
category_translation = pd.read_csv(r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive" + r'\product_category_name_translation.csv')

# Explore all tables in one shot
tables = {
    'customers': customers,
    'geolocation': geolocation,
    'orders': orders,
    'order_items': order_items,
    'order_payments': order_payments,
    'order_reviews': order_reviews,
    'products': products,
    'sellers': sellers,
    'category_translation': category_translation
}

for name, df in tables.items():
    print(f"\n{'='*40}")
    print(f"TABLE: {name}")
    print(f"Shape: {df.shape}")
    print(f"Columns: {list(df.columns)}")
    print(f"Nulls:\n{df.isnull().sum()[df.isnull().sum() > 0]}")
```

```python
# Fix 1 — Convert date columns to datetime
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
for col in date_cols:
    orders[col] = pd.to_datetime(orders[col])
```

```python
# Fix 2 — Drop rows where order was never approved
orders = orders.dropna(subset=['order_approved_at'])
```

```python
# Fix 3 — Fill missing product categories with 'unknown'
products['product_category_name'] = products['product_category_name'].fillna('unknown')
```

```python
# Fix 4 — Drop rows with missing product dimensions
products = products.dropna(subset=['product_weight_g'])
```

```python
# Fix 5 — Merge products with English category names
products = products.merge(category_translation, on='product_category_name', how='left')
products['product_category_name_english'] = products['product_category_name_english'].fillna('unknown')

print(f"Orders: {orders.shape}")
print(f"Products: {products.shape}")
```

```python
# Step 1 — Merge orders + customers
df = orders.merge(customers, on='customer_id', how='left')

# Step 2 — Merge with order_items
df = df.merge(order_items, on='order_id', how='left')

# Step 3 — Merge with products
df = df.merge(products[['product_id', 'product_category_name_english']], 
              on='product_id', how='left')

# Step 4 — Merge with sellers
df = df.merge(sellers, on='seller_id', how='left')

# Step 5 — Merge with payments
payments_agg = order_payments.groupby('order_id').agg(
    payment_value=('payment_value', 'sum'),
    payment_type=('payment_type', 'first')
).reset_index()
df = df.merge(payments_agg, on='order_id', how='left')

# Step 6 — Merge with reviews
reviews_agg = order_reviews.groupby('order_id').agg(
    review_score=('review_score', 'mean')
).reset_index()
df = df.merge(reviews_agg, on='order_id', how='left')

# Step 7 — Add useful columns
df['order_year'] = df['order_purchase_timestamp'].dt.year
df['order_month'] = df['order_purchase_timestamp'].dt.month
df['order_month_name'] = df['order_purchase_timestamp'].dt.strftime('%b')
df['delivery_days'] = (df['order_delivered_customer_date'] - 
                       df['order_purchase_timestamp']).dt.days

print("✅ Master dataframe ready!")
print(f"Shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
```

```python
# Save master dataframe
df.to_csv(r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive\olist_master.csv", index=False)
print("✅ Master CSV saved!")

# Push to SQL Server
from sqlalchemy import create_engine

engine = create_engine(
    'mssql+pyodbc://MrROBOT\\SQLEXPRESS/OlistEcommerce'
    '?driver=ODBC+Driver+17+for+SQL+Server'
    '&trusted_connection=yes'
)

df.to_sql('olist_master', engine, if_exists='replace', index=False)
print("✅ Table pushed to SQL Server!")
```

```python
import pandas as pd
from datetime import datetime

df = pd.read_csv(
    r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive\olist_master.csv"
)

# Use only delivered orders
delivered = df[df['order_status'] == 'delivered'].copy()

# Reference date = 1 day after last order
reference_date = pd.to_datetime(delivered['order_purchase_timestamp']).max() + pd.Timedelta(days=1)

print(f"Reference date: {reference_date}")
print(f"Delivered orders: {len(delivered)}")
```

```python
# Convert timestamp to datetime
delivered['order_purchase_timestamp'] = pd.to_datetime(delivered['order_purchase_timestamp'])

# Build RFM table
rfm = delivered.groupby('customer_unique_id').agg(
    recency=('order_purchase_timestamp', lambda x: (reference_date - x.max()).days),
    frequency=('order_id', 'nunique'),
    monetary=('payment_value', 'sum')
).reset_index()

print("RFM Table Sample:")
print(rfm.head())
print(f"\nShape: {rfm.shape}")
print(f"\nRFM Stats:")
print(rfm[['recency', 'frequency', 'monetary']].describe().round(2))
```

```python
# Score each customer 1-5 for R, F, M
rfm['R_score'] = pd.qcut(rfm['recency'], q=5, labels=[5,4,3,2,1])
rfm['F_score'] = pd.qcut(rfm['frequency'].rank(method='first'), q=5, labels=[1,2,3,4,5])
rfm['M_score'] = pd.qcut(rfm['monetary'], q=5, labels=[1,2,3,4,5])

# Convert to int
rfm['R_score'] = rfm['R_score'].astype(int)
rfm['F_score'] = rfm['F_score'].astype(int)
rfm['M_score'] = rfm['M_score'].astype(int)

# Combined RFM score
rfm['RFM_score'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)

# Segment customers
def segment_customer(row):
    r = row['R_score']
    f = row['F_score']
    m = row['M_score']
    
    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'
    elif r >= 3 and f >= 3 and m >= 3:
        return 'Loyal Customers'
    elif r >= 4 and f <= 2:
        return 'New Customers'
    elif r <= 2 and f >= 3:
        return 'At Risk'
    elif r <= 2 and f <= 2 and m <= 2:
        return 'Lost Customers'
    elif r >= 3 and m >= 3:
        return 'Potential Loyalists'
    else:
        return 'Needs Attention'

rfm['segment'] = rfm.apply(segment_customer, axis=1)

# Summary
segment_summary = rfm.groupby('segment').agg(
    total_customers=('customer_unique_id', 'count'),
    avg_recency=('recency', 'mean'),
    avg_frequency=('frequency', 'mean'),
    avg_monetary=('monetary', 'mean')
).round(2).reset_index().sort_values('total_customers', ascending=False)

print(segment_summary)
```

```python
# Plot 1 — Segment Distribution
plt.figure(figsize=(10,6))
colors = ['#2ecc71','#27ae60','#3498db','#f39c12','#e67e22','#e74c3c','#c0392b']
sns.barplot(data=segment_summary, x='total_customers', y='segment', palette=colors)
plt.title('Customer Segments — RFM Analysis')
plt.xlabel('Number of Customers')
plt.ylabel('Segment')
plt.tight_layout()
plt.savefig(fig_path + 'plot7_rfm_segments.png', bbox_inches='tight')
plt.show()

# Plot 2 — Avg Monetary by Segment
plt.figure(figsize=(10,6))
sns.barplot(data=segment_summary.sort_values('avg_monetary', ascending=False),
            x='avg_monetary', y='segment', palette=colors)
plt.title('Average Spend by Customer Segment')
plt.xlabel('Average Monetary Value (R$)')
plt.ylabel('Segment')
plt.tight_layout()
plt.savefig(fig_path + 'plot8_rfm_monetary.png', bbox_inches='tight')
plt.show()

# Save RFM table
rfm.to_csv(r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive\olist_rfm.csv", index=False)
print("✅ RFM plots saved and CSV exported!")
```

```python
from prophet import Prophet
import pandas as pd
import matplotlib.pyplot as plt

# Load master data
df = pd.read_csv(
    r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive\olist_master.csv"
)

# Prepare data for Prophet
# Prophet needs exactly two columns: ds (date) and y (value)
monthly_revenue = df[df['order_status']=='delivered'].groupby(
    'order_purchase_timestamp'
)['payment_value'].sum().reset_index()

# Convert to monthly
monthly_revenue['order_purchase_timestamp'] = pd.to_datetime(
    monthly_revenue['order_purchase_timestamp']
)
monthly_revenue = monthly_revenue.set_index('order_purchase_timestamp')
monthly_revenue = monthly_revenue.resample('M').sum().reset_index()

# Rename for Prophet
monthly_revenue.columns = ['ds', 'y']

print("✅ Data ready for forecasting!")
print(monthly_revenue)
```

```python
# Fix the warning + Train Prophet model
monthly_revenue = df[df['order_status']=='delivered'].groupby(
    'order_purchase_timestamp'
)['payment_value'].sum().reset_index()

monthly_revenue['order_purchase_timestamp'] = pd.to_datetime(
    monthly_revenue['order_purchase_timestamp']
)
monthly_revenue = monthly_revenue.set_index('order_purchase_timestamp')
monthly_revenue = monthly_revenue.resample('ME').sum().reset_index()
monthly_revenue.columns = ['ds', 'y']

# Remove zero revenue months (incomplete data)
monthly_revenue = monthly_revenue[monthly_revenue['y'] > 0]

print(f"Training on {len(monthly_revenue)} months of data")

# Train Prophet model
model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
    seasonality_mode='multiplicative'
)
model.fit(monthly_revenue)

# Forecast next 3 months
future = model.make_future_dataframe(periods=3, freq='ME')
forecast = model.predict(future)

# Plot forecast
fig = model.plot(forecast)
plt.title('Olist Sales Forecast — Next 3 Months')
plt.xlabel('Date')
plt.ylabel('Revenue (R$)')
plt.tight_layout()
plt.savefig(
    r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive\plot9_forecast.png",
    bbox_inches='tight'
)
plt.show()

# Show forecasted values
print("\n--- Next 3 Months Forecast ---")
print(forecast[['ds','yhat','yhat_lower','yhat_upper']].tail(3).round(2))
```

```python
# Fix — remove the Black Friday outlier and scale properly
monthly_revenue_clean = monthly_revenue.copy()

# Cap outliers — replace values above 99th percentile
cap_value = monthly_revenue_clean['y'].quantile(0.95)
monthly_revenue_clean['y'] = monthly_revenue_clean['y'].clip(upper=cap_value)

print(f"Cap value: R${cap_value:,.2f}")
print(monthly_revenue_clean)

# Retrain with clean data
model2 = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
    seasonality_mode='additive'
)
model2.fit(monthly_revenue_clean)

# Forecast next 3 months
future = model2.make_future_dataframe(periods=3, freq='ME')
forecast2 = model2.predict(future)

# Plot
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(monthly_revenue['ds'], monthly_revenue['y'], 
        'b.', label='Actual Revenue')
ax.plot(forecast2['ds'], forecast2['yhat'], 
        'r-', label='Forecast')
ax.fill_between(forecast2['ds'], 
                forecast2['yhat_lower'], 
                forecast2['yhat_upper'], 
                alpha=0.3, color='red', label='Confidence Interval')
ax.axvline(x=monthly_revenue['ds'].max(), 
           color='green', linestyle='--', label='Forecast Start')
ax.set_title('Olist Sales Forecast — Next 3 Months')
ax.set_xlabel('Date')
ax.set_ylabel('Revenue (R$)')
ax.legend()
plt.tight_layout()
plt.savefig(
    r"D:\Data_Analyst_Projects\datasets\sales_brazillian_data\archive\plot9_forecast.png",
    bbox_inches='tight'
)
plt.show()

# Show forecast values
print("\n--- Next 3 Months Forecast ---")
last3 = forecast2[['ds','yhat','yhat_lower','yhat_upper']].tail(3)
for _, row in last3.iterrows():
    print(f"{row['ds'].strftime('%B %Y')}: R${row['yhat']:,.2f} "
          f"(Range: R${row['yhat_lower']:,.2f} - R${row['yhat_upper']:,.2f})")
```
